```python
chunk_size = 500000  # np. 50 tys. na chunk
filtered_chunks = []

# Wczytuj plik w częściach
i = 0
for chunk in pd.read_csv(r'C:\Users\ulasz\OneDrive\Pulpit\studia\sem6\hurtownie danych\dane\charts.csv', chunksize=chunk_size):
    # Filtrowanie tylko top regionów
    print(f"Processing chunk {i + 1}")
    filtered = chunk[chunk['region'].isin(top_regions)]
    filtered_chunks.append(filtered)
    i += 1

# Połącz wszystkie przefiltrowane kawałki w jeden DataFrame
charts_filtered = pd.concat(filtered_chunks, ignore_index=True)
```

```python
charts = weekly_data.copy()
```

```python
weeks = charts.groupby(['region', 'chart_type', 'title'])['date'].nunique().reset_index(name='weeks_on_chart')
charts = charts.merge(weeks, on=['region', 'chart_type', 'title'], how='left')
```

```python
peak = charts.groupby(['region', 'chart_type', 'title'])['rank'].min().reset_index(name='peak_position')
charts = charts.merge(peak, on=['region', 'chart_type', 'title'], how='left')
```

```python
charts = charts.sort_values(['region', 'chart_type', 'title', 'date'])
charts['previous_rank'] = charts.groupby(['region', 'chart_type', 'title'])['rank'].shift(1)
charts['position_change'] = charts['previous_rank'] - charts['rank']
charts['position_change'] = charts['position_change'].fillna(0).astype(int)
```
