```python
import musicbrainzngs

# Inicjalizacja klienta
musicbrainzngs.set_useragent("MusicDataProject", "1.0", "youremail@example.com")

# Szukamy artysty
result = musicbrainzngs.search_artists(artist="The Weeknd", limit=1)
artist_id = result['artist-list'][0]['id']

# Pobieramy jego wydania (release-groups)
releases = musicbrainzngs.browse_release_groups(artist=artist_id, includes=["releases"], limit=5)

# Pobieramy label z konkretnego wydania
for rg in releases["release-group-list"]:
    rg_title = rg["title"]
    rg_id = rg["id"]
    print(f"\nRelease Group: {rg_title}")
    release_info = musicbrainzngs.get_release_group_by_id(rg_id, includes=["releases"])
    for release in release_info["release-group"]["release-list"]:
        if "label-info-list" in release:
            for label_info in release["label-info-list"]:
                label_name = label_info["label"]["name"]
                print(f"  Label: {label_name}")
```
